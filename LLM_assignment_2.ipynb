```python
import json
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-560m", use_fast=True)

model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-560m")

labels = ["negative", "neutral", "positive"]


def create_prompt(text, sentiment_labels):
    template = "Choose from {labels}. For example sentiment is neutral, \
    sentiment is positive, sentiment is negative. \
    {text}. sentiment is "
    labels = ", ".join(sentiment_labels)
    prompt = template.format(text=text, labels=labels)
    return prompt


def classify_text(text):
    inputs = tokenizer.encode(text, return_tensors="pt")
    result = model.generate(inputs, max_new_tokens=1)
    generated_text = tokenizer.decode(result[0], skip_special_tokens=True)
    predicted_label = generated_text.split()[-1]
    predicted_label = predicted_label.lower()
    if "negative" in predicted_label:
      label_number=0
      predicted_label="negative"
    elif "positive" in predicted_label:
      label_number=2
      predicted_label="positive"
    else:
      label_number=1
      predicted_label="neutral"
    return {"text": text, "predicted_label": predicted_label, "label_number": label_number}


texts = []
with open("/content/drive/MyDrive/Colab Notebooks/test.jsonl", "r", encoding="utf-8") as file:
    for idx, line in enumerate(file):
        data = json.loads(line)
        texts.append(data["text"])

classified_data = []
for text in texts:
    classified_data.append(classify_text(create_prompt(text, labels)))

output_file_path = "/content/drive/MyDrive/Colab Notebooks/bloom-560m_zero_shot.jsonl"
with open(output_file_path, "w", encoding="utf-8") as output_file:
    for entry in classified_data:
        output_file.write(json.dumps(entry) + "\n")

print("Results stored in bloom-560m_zero_shot.jsonl")
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Function to calculate evaluation metrics
def calculate_metrics(ground_truth_labels, predicted_labels):
    precision = precision_score(ground_truth_labels, predicted_labels, average='weighted')
    recall = recall_score(ground_truth_labels, predicted_labels, average='weighted')
    f1 = f1_score(ground_truth_labels, predicted_labels, average='weighted')
    return precision, recall, f1

# Load ground truth labels from test.json
ground_truth_labels = []
with open("/content/drive/MyDrive/Colab Notebooks/test.jsonl", "r", encoding="utf-8") as file:
    for line in file:
        data = json.loads(line)
        ground_truth_labels.append(data["label"])

# Models and their corresponding result files
models = ["bloom560m", "bloom1b1", "bloom1b7", "bloomz560m", "bloomz1b1", "bloomz1b7"]
result_files = ["bloom-560m_zero_shot.jsonl", "bloom-1b1_zero_shot.jsonl",
                "bloom-1b7_zero_shot.jsonl", "bloomz-560m_zero_shot.jsonl",
                "bloomz-1b1_zero_shot.jsonl", "bloomz-1b7_zero_shot.jsonl"]

# Initialize lists to store evaluation metrics
bloom_precisions = []
bloom_recalls = []
bloom_f1_scores = []
bloomz_precisions = []
bloomz_recalls = []
bloomz_f1_scores = []

# Load predicted labels and calculate evaluation metrics for each model
for file in result_files:
    predicted_labels = []
    with open("/content/drive/MyDrive/Colab Notebooks/" + file, "r", encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)
            predicted_labels.append(data["label_number"])

    # Convert labels to lists if they are strings
    ground_truth_labels = [int(label) for label in ground_truth_labels]
    predicted_labels = [int(label) for label in predicted_labels]

    precision, recall, f1 = calculate_metrics(ground_truth_labels, predicted_labels)
    if "bloomz" in file:
      bloomz_precisions.append(precision)
      bloomz_recalls.append(recall)
      bloomz_f1_scores.append(f1)
    else:
      bloom_precisions.append(precision)
      bloom_recalls.append(recall)
      bloom_f1_scores.append(f1)

# Plotting
x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

ax1.bar(x[:len(bloom_precisions)], bloom_precisions, width, label='Bloom', color='blue')
ax1.bar(x[len(bloom_precisions):], bloomz_precisions, width, label='Bloomz', color='red')
ax1.set_ylabel('Precision')
ax1.set_title('Precision Scores by Model')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend()

ax2.bar(x[:len(bloom_recalls)], bloom_recalls, width, label='Bloom', color='blue')
ax2.bar(x[len(bloom_recalls):], bloomz_recalls, width, label='Bloomz', color='red')
ax2.set_ylabel('Recall')
ax2.set_title('Recall Scores by Model')
ax2.set_xticks(x)
ax2.set_xticklabels(models)
ax2.legend()

ax3.bar(x[:len(bloom_f1_scores)], bloom_f1_scores, width, label='Bloom', color='blue')
ax3.bar(x[len(bloom_f1_scores):], bloomz_f1_scores, width, label='Bloomz', color='red')
ax3.set_ylabel('F1 Score')
ax3.set_title('F1 Scores by Model')
ax3.set_xticks(x)
ax3.set_xticklabels(models)
ax3.legend()

plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Function to calculate evaluation metrics
def calculate_metrics(ground_truth_labels, predicted_labels):
    precision = precision_score(ground_truth_labels, predicted_labels, average='weighted')
    recall = recall_score(ground_truth_labels, predicted_labels, average='weighted')
    f1 = f1_score(ground_truth_labels, predicted_labels, average='weighted')
    return precision, recall, f1

# Load ground truth labels from test.json
ground_truth_labels = []
with open("/content/drive/MyDrive/Colab Notebooks/test.jsonl", "r", encoding="utf-8") as file:
    for line in file:
        data = json.loads(line)
        ground_truth_labels.append(data["label"])

# Models and their corresponding result files
models = ["bloom560m", "bloom1b1", "bloom1b7", "bloomz560m", "bloomz1b1", "bloomz1b7"]
result_files = ["bloom-560m_few_shot.jsonl", "bloom-1b1_few_shot.jsonl",
                "bloom-1b7_few_shot.jsonl", "bloomz-560m_few_shot.jsonl",
                "bloomz-1b1_few_shot.jsonl", "bloomz-1b7_few_shot.jsonl"]

# Initialize lists to store evaluation metrics
bloom_precisions = []
bloom_recalls = []
bloom_f1_scores = []
bloomz_precisions = []
bloomz_recalls = []
bloomz_f1_scores = []

# Load predicted labels and calculate evaluation metrics for each model
for file in result_files:
    predicted_labels = []
    with open("/content/drive/MyDrive/Colab Notebooks/" + file, "r", encoding="utf-8") as f:
        for line in f:
            data = json.loads(line)
            predicted_labels.append(data["label_number"])

    # Convert labels to lists if they are strings
    ground_truth_labels = [int(label) for label in ground_truth_labels]
    predicted_labels = [int(label) for label in predicted_labels]

    precision, recall, f1 = calculate_metrics(ground_truth_labels, predicted_labels)
    if "bloomz" in file:
      bloomz_precisions.append(precision)
      bloomz_recalls.append(recall)
      bloomz_f1_scores.append(f1)
    else:
      bloom_precisions.append(precision)
      bloom_recalls.append(recall)
      bloom_f1_scores.append(f1)

# Plotting
x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

ax1.bar(x[:len(bloom_precisions)], bloom_precisions, width, label='Bloom', color='blue')
ax1.bar(x[len(bloom_precisions):], bloomz_precisions, width, label='Bloomz', color='red')
ax1.set_ylabel('Precision')
ax1.set_title('Precision Scores by Model')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend()

ax2.bar(x[:len(bloom_recalls)], bloom_recalls, width, label='Bloom', color='blue')
ax2.bar(x[len(bloom_recalls):], bloomz_recalls, width, label='Bloomz', color='red')
ax2.set_ylabel('Recall')
ax2.set_title('Recall Scores by Model')
ax2.set_xticks(x)
ax2.set_xticklabels(models)
ax2.legend()

ax3.bar(x[:len(bloom_f1_scores)], bloom_f1_scores, width, label='Bloom', color='blue')
ax3.bar(x[len(bloom_f1_scores):], bloomz_f1_scores, width, label='Bloomz', color='red')
ax3.set_ylabel('F1 Score')
ax3.set_title('F1 Scores by Model')
ax3.set_xticks(x)
ax3.set_xticklabels(models)
ax3.legend()

plt.tight_layout()
plt.show()
```

# Analysis

## First experiment
I tried 2 different prompts for the first experiment.

For the BLOOM models:

    "Choose from {labels}. For example sentiment is neutral, sentiment is positive, sentiment is negative. {text}. sentiment is "
  
For the BLOOMZ models:
  
    "Classify the sentiment of this text: {text}. The sentiment can be one of {labels}."
  
*  For the BLOOMZ models I tried to write a prompt with specific instructions since the BLOOMZ models are instruction tuned. This seemed to work because for the zero shot experiments the BLOOMZ models all performed much better than the BLOOM models.

* I found that in the BLOOM models the BLOOM-1b7 model performed the best. BLOOM-1b1 was the worst performing model according to the F1 score but it was the best performing model according to the precision metric. BLOOM-560m had an average performance. It did not beat BLOOM-1b7 but it did better than BLOOM-1b1.

*  In the BLOOMZ models BLOOMZ-1b7 was the best performing by a small margin which was to be expected since it is the largest among the BLOOMZ models but BLOOMZ-1b1 was very close in performance. The prompt might have also helped the increase in performance. BLOOMZ-560m perfomed slightly lesser than BLOOMZ-1b1, and BLOOMZ-1b7. All 3 models had very similar levels of performance.

*  I opted not to use any hyperparameter settings to improve the performance because I wanted to test how much the prompt affected the performce of the models. For the BLOOM models it was a real challenge coming up with an effective prompt since it is not instruction tuned and there is no guarantee that the model will generate the labels. But in the end the prompt I used seemed to work apart from a few incorrect labels being generated. The BLOOMZ models being instruction tuned was a real advantage since I could add instructions in the prompt and the model would understand the task. There were no incorrect labels predicted and for all 870 rows the model generated one of negetive, neutral, positive.

## Second Experiment

For the second experiment I opted to use the following prompts.

For the BLOOM models:

    "Choose one of {labels}. Consider the following examples: {exemplars}. {text}. Sentiment is ".

For the BLOOMZ models:

    Consider the following examples: {exemplars}. Classify the sentiment of this text: {text}. The sentiment can be one of {labels}.

*  For the second experiment I kept the code almost the same apart from the generation of the example texts. I created a separate function to extract 2 tweets of each sentiment and append it to a list. I then presented these examples to the model by including them in the prompt.

*  For the BLOOM models I thought the inclusion of the examples might help it learn what is being expected and increase the performance. But providing examples seemed to confuse the BLOOM models more than help them. The performance was across the board worse than the BLOOM models in the first experiment. Even the best performing model the BLOOM-1b7 has almost the same performance as the worst performing zero shot BLOOM model BLOOM-1b1.

* This trend seemed to be present in the BLOOMZ models as well. Each of the 3 BLOOMZ models performed worse than their zero shot counterpart. I think this same level of decrease in performance with few shot was seen by many other students as seen by the slack discussions. This was really surprising as I expected the models to perform much better given some examples to learn from but the examples seem to have confused the models more than help it learn.

*  I did experiment with the temperature hyperparameter setting to see if it changed the performance level of the BLOOM models. I kept temperature as 0.1. I found that the performance was not that much more better than opting to not use any hyperparameter setting. So I opted to not use any hyperparameter setting in the second experiment as well. I imagine that although keeping the temperature low will reduce the randomness of the prediction for the models, it still won't help the model make the correct prediction.
